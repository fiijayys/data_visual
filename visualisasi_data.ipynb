```python
fig, axes = plt.subplots(ncols=2,figsize=(8,3))

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

axes[0].plot(x1,y1)
axes[1].plot(x2,y2)

#--- Define Title ---
axes[0].set_title("Exponential Curve")
axes[1].set_title("Linear Curve")

plt.show()
```

```python
fig, axes = plt.subplots(ncols=2,figsize=(8,4))

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

axes[0].plot(x1,y1)
axes[1].plot(x2,y2)

#--- Define Title to Each Subplot---
axes[0].set_title("Exponential Curve")
axes[1].set_title("Linear Curve")

#--- Define Figure's Title---
fig.suptitle("These are My Graphs")
fig.subplots_adjust(top=0.76) #Position setting from the suptitle

plt.show()
```

```python
fig, axes = plt.subplots(ncols=2,figsize=(8,4))

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

axes[0].plot(x1,y1)
axes[1].plot(x2,y2)

axes[0].set_title("Exponential Curve")
axes[1].set_title("Linear Curve")

#Define axis label to the first subplot
axes[0].set_xlabel('X-Axis')
axes[0].set_ylabel('Y-Axis')

#Define axis label to the second subplot
axes[1].set_xlabel('X-Axis')
axes[1].set_ylabel('Y-Axis')

fig.suptitle("These are My Graphs")
fig.subplots_adjust(top=0.76)

plt.show()
```

```python
fig, ax = plt.subplots()

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

ax.plot(x1,y1, label="exponential") #You have to set the label in every chart first
ax.plot(x2,y2, label="linear")

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.set_title('Our Curves')

plt.legend() #Using this code to show the legend
plt.show()
```

```python
fig, ax = plt.subplots()

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

ax.plot(x1,y1, label="exponential")
ax.plot(x2,y2, label="linear")

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.set_title('Our Curves')

plt.legend(bbox_to_anchor=(0.75,-0.15),ncol=2)

plt.show()
```
